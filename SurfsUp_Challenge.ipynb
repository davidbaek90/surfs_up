```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd

from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==6).all()
print(june_temp)
```

```python
#  3. Convert the June temperatures to a list.
# Only adding tempreatures, excluding dates
june_results = session.query(Measurement.tobs).filter(extract('month',Measurement.date)==6).all()
print(june_results)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_results_df = pd.DataFrame(june_results)
june_results_df_rn = june_results_df.rename(columns={0: 'June Temps'})
print(june_results_df_rn)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_results_df_rn.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==12).all()
print(dec_temp)
```

```python
# 7. Convert the December temperatures to a list.
# Only adding tempreatures, excluding dates
dec_results = session.query(Measurement.tobs).filter(extract('month',Measurement.date)==12).all()
print(dec_results)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_results_df = pd.DataFrame(dec_results)
dec_results_df_rn = dec_results_df.rename(columns={0: 'December Temps'})
print(dec_results_df_rn)
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_results_df_rn.describe()
```

```python
june_results_df_rn.plot.hist(bins=20)
plt.title('June Temperature')
plt.xlim([55,85])
plt.ylim([0,400])
```

```python
june_prcp = session.query(Measurement.tobs, Measurement.prcp).filter(extract('month',Measurement.date)==6).all()
june_prcp_df = pd.DataFrame(june_prcp)
june_prcp_df_rn = june_prcp_df.rename(columns={0: 'June Temperature', 1: 'June Precipitation'})
june_prcp_df_rn.describe()
```

```python
june_precp_df_rn_plot = pd.DataFrame(june_prcp, columns=['tobs', 'prcp'])
june_precp_df_rn_plot.plot.scatter('tobs', 'prcp', figsize=(8,5))
plt.title('June Temperature vs. Precipitation')
plt.xlim([55, 85])
plt.xticks(rotation=90)
plt.ylim([0,7])
```
